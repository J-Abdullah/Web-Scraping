```python
from bs4 import BeautifulSoup
import requests
url = 'https://www.dawn.com/latest-news'

response = requests.get(url)
response.content

soup = BeautifulSoup(response.content,'html.parser')
soup.prettify()


pop_posts = soup.find_all("article", attrs={"class":"story"})

url_set = set(pop_posts) # for remove repeated links

list_of_anchor = url_set
for i in list_of_anchor:
  time = i.find('span',attrs={"class":"story__time"})
  title = i.find('h2',attrs={"class":"story__title"})
  if time:
    print("Time : ",time.text)
  else: 
    print("Time : None")
  if title:
    print("Title : ",title.text)
  else:
    print("Title : None")
  anchor = i.find('a')
  if anchor:
    print("href : ",anchor['href'])
  else:
    print("href : None")
  img= i.find('img')
  if img:
    print("img : ",img['src'])
  print("----------------------")
```
